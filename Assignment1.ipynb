```python
# Assignment 1, Problem 1
#---------------------------------------------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from my_functions import *
```

```python
df = pd.read_csv('./HW1.csv')
df.head()
m = len(df)
m
```

```python
df
```

```python
# Seperate features and labels
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
Y = df.values[:,3]
m = len(Y)
n = len(X1)

print('X1 = ', X1[: 5])
print('X2 = ', X2[: 5])
print('X3 = ', X3[: 5])
print('Y = ', Y[: 5])
print('m = ',m)
print('n = ',n)
```

```python
# X1 Scatter Plot
plt.scatter(X1,Y,color='b')
plt.grid()
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('X1 -> Y Scatter Plot');
```

```python
# Build linear regression model
X1_0 = np.ones((m,1))                    # Bias column (same size as data points)
X1_0[:5]
```

```python
X1_1 = X1.reshape(m,1)                   # Reshaping X1 to fit matrix operations
X1_1[:10]
```

```python
X1_feat = np.hstack((X1_0,X1_1))          # Concatonate bias and training data horizontally
X1_feat[:5]
```

```python
theta = np.zeros(2)                       # Create 2-column vector for theta values
theta
```

```python
# Computing J(theta_0, theta_1)
iterations = 1500;
alpha = 0.05;

theta1, cost_history = gradient_descent(X1_feat,Y,theta,alpha,iterations,m)
print('Final value of [Theta_0, Theta_1] = ', theta1)
print('cost_history =', cost_history)
```

```python
# X1 Linear Regression Graph
plt.scatter(X1_feat[:,1], Y, color='b', label='Training Data')
plt.plot(X1_feat[:,1], X1_feat.dot(theta1), color='r', label='Linear Regression')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression X1')
plt.legend(); plt.grid()
```

```python
# X1 Loss Graph
plt.plot(range(1, iterations + 1), cost_history, color='k')
plt.grid()
plt.xlabel('Iterations')
plt.ylabel('Loss (J)')
plt.title('X1 Gradient Descent');
```

```python
# X2 Scatter Plot
plt.scatter(X2,Y,color='g')
plt.grid()
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('X2 -> Y Scatter Plot');
```

```python
# Building X2 Linear Regression Model
X2_0 = np.ones((m,1))
X2_1 = X2.reshape(m,1)
X2_feat = np.hstack((X2_0,X2_1))
```

```python
theta = np.zeros(2)
```

```python
# Calculating J(theta_0,theta_1)
iterations = 1500;
alpha = 0.05;

theta2, cost_history = gradient_descent(X2_feat,Y,theta,alpha,iterations,m)
```

```python
# X2 Linear Regression Plot
plt.scatter(X2_feat[:,1], Y, color='g', label='Training Data')
plt.plot(X2_feat[:,1], X2_feat.dot(theta2), color='r', label='Linear Regression')
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('Linear Regression of X2 -> Y Scatter')
plt.grid(); plt.legend();
```

```python
# X2 Loss Graph
plt.plot(range(1, iterations + 1), cost_history, color='k')
plt.xlabel('Iterations')
plt.ylabel('Loss (J)')
plt.title('X2 Loss Graph');
plt.grid()
```

```python
# X3 Scatter Plot
plt.scatter(X3, Y, color='r')
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('X3 -> Y Scatter Plot')
plt.grid()
```

```python
# Building X3 Linear Regression Model
X3_0 = np.ones((m,1))
X3_1 = X3.reshape(m,1)
X3_feat = np.hstack((X3_0,X3_1))
theta = np.zeros(2)
```

```python
# Calculating Cost Function (J)
iterations = 1500;
alpha = 0.05;
theta3, cost_history = gradient_descent(X3_feat,Y,theta,alpha,iterations,m)
```

```python
# X3 Linear Regression Scatter Plot
plt.scatter(X3_feat[:,1], Y, color='r', label='Training Data')
plt.plot(X3_feat[:,1],X3_feat.dot(theta3), color='b', label='Linear Regression')
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('Linear Regression of X3 -> Y Scatter Plot')
plt.grid(); plt.legend();
```

```python
# X3 Loss Graph
plt.plot(range(1, iterations + 1), cost_history, color='k')
plt.title('X3 Loss Graph')
plt.xlabel('Iterations')
plt.ylabel('Loss (J)')
plt.grid();
```

```python
# Remove output column for easy-access
df1 = df.iloc[:,:-1]
df1
```

```python
# Init X1, X2, X3
X1 = df1.values[:,0]
X2 = df1.values[:,1]
X3 = df1.values[:,2]
# Add Bias
X0 = np.ones((m,1))
# Ensure X's are in right shape
X1 = X1.reshape(m,1)
X2 = X2.reshape(m,1)
X3 = X3.reshape(m,1)
# Concatonate
X_full = np.hstack((X0,X1,X2,X3))
```

```python
theta = np.zeros(4) # 3 variables --> 4 coefficients
theta
```

```python
# Calculating Cost Function (J)
iterations = 1500; # Arbritrary
alpha = 0.05;      # Between 0.01 and 0.1
theta, cost_history = gradient_descent(X_full,Y,theta,alpha,iterations,m)
theta
```

```python
plt.scatter(X_full[:,1], Y, color='r', label='Training Data X1')
plt.scatter(X_full[:,2], Y, color='g', label='Training Data X2')
plt.scatter(X_full[:,3], Y, color='b', label='Training Data X3')

# X1 Line with y-intercept
# plt.plot(X_full[:,1],X_full[:,0].dot(theta[0])+X_full[:,1].dot(theta[1]),color='k',label='Linear Regression (Best Fit)');

# X2 Line with y-intercept
# plt.plot(X_full[:,2],X_full[:,0].dot(theta[0])+X_full[:,2].dot(theta[2]),color='k',label='Linear Regression (Best Fit)');

# X3 Line with y-intercept (CHOSEN BEST FIT)
plt.plot(X_full[:,3],X_full[:,0].dot(theta[0])+X_full[:,3].dot(theta[3]),color='k',label='Linear Regression (Best Fit)');

plt.grid(); plt.legend();
plt.xlabel('X'); plt.ylabel('Y');
plt.title('Multiple Linear Regression');
```

```python
plt.plot(range(1, iterations + 1), cost_history, color='k');
plt.grid();
plt.xlabel('Iterations'); plt.ylabel('Cost (J)');
plt.title('Cost vs. Iterations');
```
